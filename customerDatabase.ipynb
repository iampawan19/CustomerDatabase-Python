```python
import pandas as pd

# Import data from Excel files
Raw_file = pd.read_excel('C:/Users/pawan/Python Projects/Customer-Python/Iconic_01 Jan to 31 March 2021_raw file.xlsx')
Outlet_file = pd.read_excel('C:/Users/pawan/Python Projects/Customer-Python/Iconic Outlets (1).xlsx')
Dummy_file = pd.read_excel('C:/Users/pawan/Python Projects/Customer-Python/Customer_Dummy_File.xlsx')



# Create a new column "Reg_id" in the first position of Raw_file
Raw_file.insert(0, 'Reg_id', None)  # You can set the default value to None or any other value

Raw_file.head()



# Assuming 'column-x' is the column in both files that you want to match on
# Add an extra column in df_a to store the values from df_b based on the vlookup
Raw_file['Reg_id'] = Raw_file['branch_code'].map(Outlet_file.set_index('Branch Code')['Registration ID'])

# Convert the 'Date' column to datetime format if it's not already in datetime
Raw_file['bill_date'] = pd.to_datetime(Raw_file['bill_date'], errors='coerce')

# Display the start date and end date from the 'Date' column
start_date = Raw_file['bill_date'].min()
end_date = Raw_file['bill_date'].max()

print(f"Start Date: {start_date}")
print(f"End Date: {end_date}")

# Convert 'Reg_id' column to integer
Raw_file['Reg_id'] = Raw_file['Reg_id'].astype('Int64', errors='ignore')

# Drop rows with NA values in the 'Reg_id' column
Raw_file.dropna(subset=['Reg_id'], inplace=True)

# Display the updated DataFrame
Raw_file.describe


# Create a new column "Reg_id" in the first position of Raw_file
Raw_file.insert(1, 'Unique_id', None)  # You can set the default value to None or any other value


# Convert 'bill_no' column to string
Raw_file['bill_no'] = Raw_file['bill_no'].astype(str)

# Update the existing 'Unique_id' column by concatenating 'Reg_id,' 'bill_no,' and 'bill_date'
Raw_file['Unique_id'] = (
    Raw_file['Reg_id'].astype(str) +
    '_' +
    Raw_file['bill_no'] +
    '_' +
    Raw_file['bill_date'].dt.strftime('%Y%m%d')
)

# Display the updated DataFrame
Raw_file.head(100)

# Filter 'customer_mob_no' column
Raw_file['customer_mob_no'] = Raw_file['customer_mob_no'].astype(str)  # Convert to string for string operations
Raw_file['customer_mob_no'] = Raw_file['customer_mob_no'].apply(lambda x: x if x.isdigit() and len(x) == 10 and x[0] in ('6', '7', '8', '9') else '')

# Display the updated DataFrame
Raw_file.head(100)


# Assuming df is your DataFrame
selected_columns = Raw_file[['customer_mob_no']]
print(selected_columns)





# Filter 'customer_mob_no' column
filtered_data = Raw_file[Raw_file['customer_mob_no'].str.match('^[6-9]\d{9}$')]

# Display the filtered DataFrame
filtered_data.head()
```
